# Correlação

## Importações

```python
import numpy as np
import pandas as pd
import seaborn as sns
import math
```

## Base de dados

```python
tamanho = np.array([30, 39, 49, 60])
preco = np.array([57000, 69000, 77000, 90000])
```

```python
dataset = pd.DataFrame({'tamanho': tamanho, 'preco': preco})
dataset
```

```python
sns.scatterplot(x=tamanho, y=preco)
```

## Correlação - cálculo com numpy e pandas

```python
np.corrcoef(tamanho, preco)
```

```python
dataset.corr()
```
