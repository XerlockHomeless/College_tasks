```python
# Regressão Linear Simples
# Importando as bibliotecas
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importando o dataset
datasets = pd.read_csv('Salary_Data.csv')
```

```python
# lista os 5 primeiros registros do dataset
datasets.head()
```

```python
# lista a estrutura do dataset (linhas X colunas)
datasets.shape
```

```python
# seleciona as variáveis X e Y do dataset

# método iloc é utilizado para seleção baseada em índices inteiros no Pandas
# a) :: Seleciona todas as linhas do DataFrame.
# b) :-1: Seleciona todas as colunas, exceto a última.
# c) .values: Retorna os valores correspondentes como um array NumPy.
X = datasets.iloc[:, :-1].values
# Portanto, esta linha extrai todas as linhas e todas as colunas, exceto a última,
# do DataFrame chamado "datasets" e armazena esses valores na variável X.
# Isso é comumente feito quando você está lidando com features (características)
# em um conjunto de dados.


# a) :: Seleciona todas as linhas do DataFrame.
# b) 1: Seleciona apenas a segunda coluna.
# c) .values: Retorna os valores correspondentes como um array NumPy.
Y = datasets.iloc[:, 1].values
# Todos os valores da segunda coluna do DataFrame datasets são extraídos e armazenados na variável Y.
# Isso é geralmente feito quando você está lidando com a variável de destino ("target")
# em um problema de aprendizado supervisionado.
```

```python
# imprime a variável X
print(X)
```

```python
# imprime a variável Y
print(Y)
```

```python
# Dividindo (splitting) o dataset em conjunto de Treinamento e de Teste
from sklearn.model_selection import train_test_split
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X, Y, test_size = 1/3, random_state = 0)
```

```python
# Fitting Simple Linear Regression (SLR) to the training set
# Ajustando (fitting) um caso de SLR ao conjunto de treinamento
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_Train, Y_Train)
```

```python
# Prevendo o resultado do conjunto de Teste
Y_Pred = regressor.predict(X_Test)
```

```python
# Visualizando os resultados do conjunto de Treinamento
plt.scatter(X_Train, Y_Train, color = 'red')
plt.plot(X_Train, regressor.predict(X_Train), color = 'blue')
plt.title('Salary vs Experience  (Training Set)')
plt.xlabel('Years of experience')
plt.ylabel('Salary')
plt.show()
```

```python
# Visualising the Test set results
# Visualizando os resultados do conjunto de Teste
plt.scatter(X_Test, Y_Test, color = 'red')
# plt.plot(X_Train, regressor.predict(X_Train), color = 'blue')
plt.plot(X_Test, regressor.predict(X_Test), color = 'blue')
plt.title('Salary vs Experience  (Test Set)')
plt.xlabel('Years of experience')
plt.ylabel('Salary')
plt.show()
```
